# A diminishing Democrat presence in Nebraska

This project uses ggplot to visualize a diminishing Democrat presence in Nebraska when comparing the 2016 election to previous elections.

```r
library(ggplot2)
library(reshape2)
library(scales)
```

Import turnout data.

```r
turnout <- read.csv("Data/turnout.csv")
```

```r
head(turnout)
```

Create a bar chart, placing registered voter numbers next to actual voter numbers.

```r
df <- ggplot(turnout, aes(x=party, y=value, fill=party, alpha=variable)) + geom_bar(colour="black", stat="identity", position=position_dodge()) + scale_x_discrete(limits=c("Republican","Democrat","Libertarian")) + scale_y_continuous(labels = comma) + scale_fill_manual(values=c("blue","yellow","red"))+ scale_alpha_manual(values=c(1, 0.5)) + ggtitle("Registered Voters vs. Votes Counted")+ ylab("Number of Voters") + guides(fill=FALSE) + theme_bw()
```

Create a facet grid based on the election year.

```r
df + facet_grid(. ~ year)
```

Save it out to a PDF for editing in Illustrator.

```r
ggsave(file="Images/turnout.pdf", width=40, height=15)
```
